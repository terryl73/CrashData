### Perform API Calls
* Pull full list of crashes in a given state for years 2012-2017.


```python
#Crash List Information data pull from Crashstats.

#Two Varaibles, state and year
state_id=9
#CrashListDF = pd.DataFrame(columns=["CountyName", "CrashDate", "Fatals", "Peds", "Persons", "St_Case", "State", "StateName", "TotalVehicles"])

#CountyList = []
#DateList = []

def Pull_Crash(state_id, CrashYear):
    Target1 = "https://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCaseList?states="+str(state_id)
    Target2 = "&fromYear="+str(CrashYear)+"&toYear="+str(CrashYear)+"&minNumOfVehicles=1&maxNumOfVehicles=50&format=json"
    Target_Url = Target1 + Target2

#Assign Json Results to DataFrame
    response = requests.get(Target_Url).json()
    CrashRaw = pd.DataFrame.from_dict(response["Results"][0])
    CrashRaw["Year"] = CrashYear
    return CrashRaw
    

Crash2012 = Pull_Crash(9,2012)
print(2012)
print(Crash2012.head())

Crash2013 = Pull_Crash(9,2013)
#print(2013)
#print(Crash2013.head())

Crash2014 = Pull_Crash(9,2014)
#print(2014)
#print(Crash2014.head())

Crash2015 = Pull_Crash(9,2015)
#print(2015)
#print(Crash2015.head())

Crash2016 = Pull_Crash(9,2016)
#print(2016)
#print(Crash2016.head())

Crash2017 = Pull_Crash(9,2017)
#print(2017)
#print(Crash2017.head())
 
DFList = [Crash2012, Crash2013, Crash2014, Crash2015, Crash2016, Crash2017]    

CrashListDF = pd.concat(DFList)

CrashListDF.describe()

#CrashListDF = pd.DataFrame()
#CrashListDF["CountyName"] = CountyList
#CrashListDF["CrashDate"] = DateList
        
#CrashListDF.head()
#CrashListDF.describe()
    
```

### Modify Data into more usable formats
* Trim County Name
* Trim Date String

```python
#Modify County Name
CountyString = CrashListDF['CountyName']
AddCounty = []

for county in CountyString:
    TempCounty = county.split("(")[0].rstrip()
    AddCounty.append(TempCounty)

CrashListDF['CountyName'] = AddCounty
```

```python
#Modify Crash Date
DateString = CrashListDF['CrashDate']

AddDate = []

for date in DateString:
    TempDate = date.split("(")[1].split(")")[0]
    #TempDate = TempDate.split("-")[0]
    TempDate = TempDate[:10]
    AddDate.append(TempDate)

CrashListDF['CrashDate'] = AddDate



CrashListDF.head()
```
